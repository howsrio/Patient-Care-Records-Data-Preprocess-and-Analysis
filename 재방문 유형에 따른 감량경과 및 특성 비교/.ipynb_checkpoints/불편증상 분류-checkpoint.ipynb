```python
import re

def extract_symptom(text):
    # 정규식 패턴 정의
    pattern = re.compile(r'(불편사항|불편증상)\s*:\s*(.*?)(?:\n|$)')
    
    # 패턴 매칭
    matches = pattern.findall(text)
    
    # 결과 리스트 생성
    results = []
    
    for match in matches:
        symptom = match[1].strip()  # 패턴의 두 번째 그룹에 해당하는 증상 부분
        if symptom.startswith('(') and symptom.endswith(')'):
            # 괄호를 제거하고 문자열 추가
            results.append(symptom[1:-1].strip())
        else:
            # 줄바꿈까지의 문자열 추가
            results.append(symptom)
    
    if len(results) == 0:
        return '없음'
    else:
        return results[0]

# 예시 텍스트
text = """
□ 복용량(아침/점심/저녁) : 2정 (or 1/2포),2정 (or 1/2포),2정 (or 1/2포)

□ 불편사항 : 두근거림,어지러움,입마름, 갈증,울렁거림,설사

□ 불편증상 : 두근거림,어지러움,입마름, 갈증,울렁거림,설사

□ 일상생활 불가능 : 아니요

□ 식사 횟수(총/죽식/간식) : 3회,1회,0회
"""

# 함수 실행
results = extract_symptom(text)
print(results)  # ['두근거림,어지러움,입마름, 갈증,울렁거림,설사', '두근거림,어지러움,입마름, 갈증,울렁거림,설사']

```

```python
symptom_dict={}
first_dict={}
symptom_list= []
for idx in progress_info_df.index:
    id = progress_info_df.loc[idx,'PatientID']
    region =progress_info_df.loc[idx,'Region']
    index_info = progress_info_df.loc[idx,['Age','Coffee','Alcohol','한방 경험','한방 불편증상 여부', '양방 경험', '양방 불편증상 여부']]
    N = int(progress_info_df.loc[idx, 'N'])
    index_N1, index_N2 = col_list.index('Weight_'+str(N)), col_list.index('ScheduleID_'+str(N+1))
    imsi = basis[(basis['Region_1']==region) & (basis['PatientID'] == id)].iloc[:,index_N1:index_N2]
    imsi_col = ['_'.join(col.split('_')[:-1]) for col in imsi.columns]
    imsi.columns = imsi_col
    imsi = imsi.reset_index(drop = True)
    if imsi.shape[0]==0:
        continue
    idx_first = pd.concat([index_info, imsi.loc[0]])
   
    symptom_idx_dict = {}
    for i in range(N+1, 25):
        ProgressNote = 'ProgressNote_'+ str(i)
        text = progress_info_df.loc[idx, ProgressNote]
        if not pd.isna(text):
            symptom = extract_symptom(text)
            symptom_idx_dict[i] = symptom
            symptom_list.append(symptom)
    symptom_dict[idx] = symptom_idx_dict
    first_dict[idx] = idx_first
```

```python
from collections import Counter

def count_elements(lst):
    # Counter를 사용하여 리스트 원소의 개수 세기
    element_count = Counter(lst)
    sorted_dict = dict(sorted(dict(element_count).items(), key=lambda item: item[1], reverse=True))
    # Counter 객체를 딕셔너리로 변환하여 반환
    return sorted_dict
def split_and_strip(lst):
    # 결과를 저장할 리스트
    result = []
    
    # 리스트의 각 원소에 대해 반복
    for item in lst:
        # 각 원소를 ','로 분리한 후 공백 제거
        parts = [word.strip()  for part in item.split(',') for word in part.split()]
        # 결과 리스트에 추가
        result.extend(parts)
    
    return result
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_selection import RFE, SelectFromModel

# 예제 데이터 프레임 생성
# FirstVisitDf = pd.read_csv('your_data.csv')  # 데이터프레임 로드

# 예제 데이터프레임에서 설명변수(X)와 목표변수(Y) 분리
X = FirstVisitDf.iloc[:,:-11]
Y = FirstVisitDf.iloc[:,-11:]

# 함수 정의: 각 y_col에 대해 특성 중요도를 평가하고 최적의 변수 조합을 찾기
def evaluate_feature_importance_and_optimize(X, Y):
    for y_col in Y.columns:
        print(f'symptom type: {y_col}')
        y = Y[y_col]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
        
        # Random Forest 모델을 사용하여 특성 중요도 평가
        model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
        model_rf.fit(X_train, y_train)
        
        # 특성 중요도 출력
        importances = model_rf.feature_importances_
        feature_importance_df = pd.DataFrame({
            'Feature': X.columns,
            'Importance': importances
        }).sort_values(by='Importance', ascending=False)
        print("Feature Importances:\n", feature_importance_df)
        
        # RFE를 사용하여 최적의 특성 조합 찾기
        rfe = RFE(estimator=LogisticRegression(), n_features_to_select=5)
        rfe.fit(X_train, y_train)
        
        # 선택된 특성
        selected_features = X.columns[rfe.support_].tolist()
        print("Selected Features by RFE:", selected_features)
        
        # 선택된 특성으로 모델 재학습 및 평가
        X_train_rfe = rfe.transform(X_train)
        X_test_rfe = rfe.transform(X_test)
        
        # 로지스틱 회귀
        model_lr = LogisticRegression()
        model_lr.fit(X_train_rfe, y_train)
        y_pred_lr = model_lr.predict(X_test_rfe)
        print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
        print(classification_report(y_test, y_pred_lr))
        
        # Random Forest
        model_rf.fit(X_train_rfe, y_train)
        y_pred_rf = model_rf.predict(X_test_rfe)
        print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
        print(classification_report(y_test, y_pred_rf))
        
        # SVM
        model_svm = SVC(kernel='linear')
        model_svm.fit(X_train_rfe, y_train)
        y_pred_svm = model_svm.predict(X_test_rfe)
        print("Support Vector Machine Accuracy:", accuracy_score(y_test, y_pred_svm))
        print(classification_report(y_test, y_pred_svm))

# 함수 호출: 특성 중요도 평가 및 최적의 변수 조합 찾기
evaluate_feature_importance_and_optimize(X, Y)
```

```python
import numpy as np

# 예시 혼동 행렬
cm = np.array([[50, 10], [5, 35]])  # 혼동 행렬 예시

def print_confusion_matrix(cm, class_names):
    """
    혼동 행렬을 콘솔에 보기 좋게 출력하는 함수
    """
    # 혼동 행렬의 행렬을 텍스트로 변환
    print("Confusion Matrix:")
    print(" " * 12, end="")
    for name in class_names:
        print(f"{name:>8}", end=" ")
    print()
    
    for i, row in enumerate(cm):
        print(f"{class_names[i]:<12}", end=" ")
        for value in row:
            print(f"{value:>8}", end=" ")
        print()

# 클래스 이름 지정
class_names = ['Class 0', 'Class 1']

# 혼동 행렬 출력
print_confusion_matrix(cm, class_names)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import torch
import torch.nn as nn
import torch.optim as optim
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 준비
X = FirstVisitDf.iloc[:, :-11].values
Y = FirstVisitDf.iloc[:, -11:].values
scaler = StandardScaler()
X = scaler.fit_transform(X)
col_names = FirstVisitDf.iloc[:, -11:].columns.tolist()

# SimpleNN 모델 정의
class SimpleNN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers):
        super(SimpleNN, self).__init__()
        layers = []
        layers.append(nn.Linear(input_dim, hidden_dim))
        layers.append(nn.ReLU())
        for _ in range(num_layers - 1):
            layers.append(nn.Linear(hidden_dim, hidden_dim))
            layers.append(nn.ReLU())
        layers.append(nn.Linear(hidden_dim, output_dim))
        layers.append(nn.Sigmoid())
        self.network = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.network(x)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 모델 저장 딕셔너리
model_dict = {}
best_accuracy_dict = {}
#Y = Y[:, 1:3]  # Assuming we want to evaluate only the first two columns

for y_col_idx in range(Y.shape[1]):
    y = Y[:, y_col_idx]
    print(f'\nSymptom type: {col_names[y_col_idx]}\n')
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1111, stratify=y)
    
    X_train = torch.tensor(X_train, dtype=torch.float32).to(device)
    y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1).to(device)
    X_test = torch.tensor(X_test, dtype=torch.float32).to(device)
    y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1).to(device)
    
    # 모델 설정
    input_dim = X_train.shape[1]
    hidden_dim = 64
    output_dim = 1
    early_stopping_threshold = 0.60  # 성능이 일정 수준 이하로 떨어지면 멈춤
    
    best_accuracy = 0.0  # 가장 좋은 F1 점수 초기값
    
    for num_layers in range(4, 9):
        model = SimpleNN(input_dim, hidden_dim, output_dim, num_layers).to(device)
        criterion = nn.BCELoss()
        optimizer = optim.Adam(model.parameters(), lr=0.0001)
        
        # 훈련
        num_epochs = 200000
        for epoch in range(num_epochs):
            model.train()
            optimizer.zero_grad()
            outputs = model(X_train)
            loss = criterion(outputs, y_train)
            loss.backward()
            optimizer.step()
            
            # 에폭 10000번마다 현재 결과 출력
            if (epoch + 1) % 50000 == 0:
                with torch.no_grad():
                    model.eval()
                    y_pred = model(X_test)
                    y_pred_class = (y_pred > 0.5).float()
                    accuracy = (y_pred_class == y_test).float().mean().item()
                    precision = precision_score(y_test.cpu(), y_pred_class.cpu())
                    recall = recall_score(y_test.cpu(), y_pred_class.cpu())
                    f1 = f1_score(y_test.cpu(), y_pred_class.cpu())
                    auc = roc_auc_score(y_test.cpu(), y_pred.cpu())
                    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}, AUC: {auc:.4f}')
                    
                    # 조기 종료 조건 체크
                    if accuracy < early_stopping_threshold:
                        print(f'Early stopping at epoch {epoch + 1} due to accuracy below threshold.')
                        break
        
        # 최종 평가 및 모델 저장
        model.eval()
        with torch.no_grad():
            y_pred = model(X_test)
            y_pred_class = (y_pred > 0.5).float()
            accuracy = (y_pred_class == y_test).float().mean().item()
            precision = precision_score(y_test.cpu(), y_pred_class.cpu())
            recall = recall_score(y_test.cpu(), y_pred_class.cpu())
            f1 = f1_score(y_test.cpu(), y_pred_class.cpu())
            auc = roc_auc_score(y_test.cpu(), y_pred.cpu())
            cm = confusion_matrix(y_test.cpu(), y_pred_class.cpu())
            print(f'\nLayers: {num_layers}, Final Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}, AUC: {auc:.4f}')
            print_confusion_matrix(cm, class_names)
            print()

            # 성능이 더 좋은 경우에만 모델을 저장
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                model_dict[f'model_{col_names[y_col_idx]}'] = model
                best_accuracy_dict[f'model_{col_names[y_col_idx]}'] = best_accuracy
                print(f'New best model for {col_names[y_col_idx]} with acc Score {best_accuracy:.4f}')
                print('-'*100)
```
