```python
from BasisFilter import BasisFilter
```

```python
BF = BasisFilter()
basis = BF.load_basis()
```

```python
for i in basis['ProgressNote_2'][100:200]:
    if isinstance(i,float):
        continue
    sentances = extract_sentences_with_keyword(i,'불편')
    for sentance in sentances:
        print(sentance)
    print()
    print('-'*50)
    print()
```

```python
def extract_sentences_with_keyword(text, keyword):
    sentences = text.split('\n')
    keyword_sentences = [sentence for sentence in sentences if keyword in sentence]
    
    return keyword_sentences

```
