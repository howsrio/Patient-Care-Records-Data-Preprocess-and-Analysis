```python
import pandas as pd
import os
```

```python
region_list = ['bundang','hongdae','bucheon','incheon','jamsil','busan','gangnam']
personal_dict ={}
path = os.getcwd()
path = os.path.dirname(path)
path = os.path.dirname(path)
path
```

```python
byPatient = pd.read_csv(f'{path}\\MergeRegionData\\byPatientID.csv')
```

```python
drop_list = ['PatientCellphone', 'PatientPhone','PatientCellphone2','PatientPhone2','GuardianNoID','Guardian',
 'GuardianCellphone',
 'GuardianPhone',
 'PatientEmail',
 'PatientWeb','PatientMajAreaCC','PatientJob','PatientZipCode2',
 'PatientAddr21',
 'PatientAddr22',
 'PatientZipCode3',
 'PatientAddr31',
 'PatientAddr32','PatientNickName','PatientIntroRel','Discarded',
 'UnpaidAmt','LunarBirthday','PhoneMemo',
 'PhoneMemo2',
 'FtFlag',
 'Nation',
 'PatientID_ORG',
 'Guest','Encrypted']
```

```python
for region in region_list:
    personal_df = pd.read_csv(f'{path}\\personal_data\\{region}.csv', encoding ='utf-8')
    personal_df = personal_df.drop(columns=drop_list)
    personal_df['Region_1'] = region

    personal_dict[region] = personal_df
```

```python
personal_dict['bundang']
```

```python
for region in region_list:
    df = personal_dict[region]
    if region == region_list[0]:
        concat_df = df
    else:
        concat_df = pd.concat([concat_df,df])
all_personal_df = concat_df
```

```python
all_personal_df.shape
```

```python
byPatient.shape
```

```python
df = pd.merge(all_personal_df, byPatient, on = ['PatientID','Region_1'], how = 'inner')
```
