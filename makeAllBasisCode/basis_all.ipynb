```python
import re

# 체중 정보 추출 함수 정의
def extract_weight_info(texts):
    
    # 정규 표현식 패턴 정의
    pattern1 = re.compile(r'키,\s*체중\s*:\s*\d{2,3}(?:\.\d{1,2})?\s+(\d{2,3}(?:\.\d{1,2})?)')
    pattern2 = re.compile(r'체중\s*:\s*(\d{2,3})')
    pattern3 = re.compile(r'키/체중/BMI:\d{2,3}\s*/\s*(\d{2,3})')
    if not isinstance(texts, str):
        print('문자열이 아닙니다.')
        return 0 # 문자열이 아닌 경우 건너뜁니다.
    texts = texts.split('\n')
    
    for text in texts:
        
        # 패턴 1: '키, 체중 : 163 67'
        match1 = pattern1.search(text)
        if match1:
            print(match1.group(1))
            print(text)
            weight = match1.group(1)
            return weight
            
        match3 = pattern3.search(text)
        if match3:
            print(match3.group(1))
            print(text)
            weight = match3.group(1)
            return weight
        
        # 패턴 2: '체중 : 67'
        match2 = pattern2.search(text)
        if match2:
            print(match2.group(1))
            print(text)
            weight = match2.group(1)
            return weight
        
        # 패턴 3: '키/체중/BMI:160 / 61 / 과체중'
            
    


```
