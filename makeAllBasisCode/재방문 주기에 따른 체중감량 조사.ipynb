```python
import os
import pandas as pd
import numpy as np
import copy
from datetime import datetime, timedelta
```

```python
BF= BasisFilter()
```

```python
BF.load_basis()
```

```python
basis = BF.basis
```

```python
first_df =BF.first_visit_filter(basis)
```

```python
notfirst_df = basis[~basis.index.isin(first_df.index)]
```

```python
first_in2_dict = BF.make_inbody2up_dict(first_df)
notfirst_in2_dict = BF.make_inbody2up_dict(notfirst_df)
```

```python
first_return_dict = BF.make_return_dict(basis, first_in2_dict,[50,120])
```

```python
day_dict = {}
for i in range(59,120):
    return_dict = BF.make_return_dict(basis, first_in2_dict,[i,i+1])
    BF.make_medicine_dict(basis, first_in2_dict, return_dict)
    day_dict[i]= BF.record_dict
```

```python
for idx in list(day_dict[60].keys()):
    print(first_return_dict[idx])
    print( day_dict[60][idx])
```
