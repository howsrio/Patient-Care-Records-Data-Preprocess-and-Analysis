```python
import re

def diet_patterns_11(text):
    # 정규 표현식을 사용하여 패턴 사이의 텍스트 추출
    pattern = r'ㆍ 다이어트 경험:\s*(.*?)\s*■'

    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        return match.group(1).strip()
    else:
        return None
def diet_patterns_12(text):
    # 정규 표현식을 사용하여 패턴 사이의 텍스트 추출
    pattern = r'ㆍ 다이어트 경험:\s*(.*?)\s*\[2\]'
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        return match.group(1).strip()
    else:
        return None
        
def diet_patterns_21(text):
    # 정규 표현식을 사용하여 패턴 사이의 텍스트 추출
    pattern = r'■ 다이어트 약 복용 경험\s*(.*?)\s*■'
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        return match.group(1).strip()
    else:
        return None
def diet_patterns_22(text):
    # 정규 표현식을 사용하여 패턴 사이의 텍스트 추출
    pattern = r'■ 다이어트 약 복용 경험\s*(.*?)\s*\[2\]'
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        return match.group(1).strip()
    else:
        return None

def diet_patterns(text):
    pattern11= diet_patterns_11(text)
    pattern12= diet_patterns_12(text)
    pattern21= diet_patterns_21(text)
    pattern22= diet_patterns_22(text)

    if pattern11 == None and pattern12 == None:
        pattern1 = None
    elif pattern11 == None or pattern11 == r'\s*':
        pattern1 = pattern12
    elif pattern12 == None or pattern12 == r'\s*':
        pattern1 = pattern11
    else:
        if pattern11 in pattern12:
            pattern1 = pattern11
        elif pattern12 in pattern11:
            pattern1 = pattern12
    
    if pattern21 == None and pattern22 == None:
        pattern2 = None
    elif pattern21 == None or pattern21 == r'\s*':
        pattern2 = pattern22
    elif pattern22 == None or pattern22 == r'\s*':
        pattern2 = pattern21
    else:
        if pattern21 in pattern22:
            pattern2 = pattern21
        elif pattern22 in pattern21:
            pattern2 = pattern22
    
    return pattern1, pattern2

def diet_detail(text):
    if text == None:
        hanbang, yangbang = [None, None, None],[None, None, None]
    elif '無' in text:
        return [0, None, None],[ 0, None, None]
    elif True:
        if '양방'  in text and '한방' in text:
            #print('양한 모두 있는 경우:',text)
            symptom = find_first_uncomfortable_symptom(text, '한방')
            if symptom == None:
                hanbang = [1, None, symptom]
            elif ((('없음' == symptom or 'x' == symptom) or '무' == symptom) or 'X' == symptom) or '-':
                hanbang = [1, 0, symptom]
            else:
                hanbang = [1, 1, symptom]
            symptom = find_first_uncomfortable_symptom(text, '양방')
            if symptom == None:
                yangbang = [1, None, symptom]
            elif ((('없음' == symptom or 'x' == symptom) or '무' == symptom) or 'X' == symptom) or '-':
                yangbang = [1, 0, symptom]
            else:
                yangbang = [1, 1, symptom]
        
        elif '양방' in text:
            symptom = find_first_uncomfortable_symptom(text, '양방')
            if symptom == None:
                yangbang = [1, None, symptom]
            elif ((('없음' == symptom or 'x' == symptom) or '무' == symptom) or 'X' == symptom) or '-':
                yangbang = [1, 0, symptom]
            else:
                yangbang = [1, 1, symptom]
            hanbang = [0, None, None]
            

        elif '한방' in text:
            symptom = find_first_uncomfortable_symptom(text, '한방')
            if symptom == None:
                hanbang = [1, None, symptom]
            elif ((('없음' == symptom or 'x' == symptom) or '무' == symptom) or 'X' == symptom) or '-':
                hanbang = [1, 0, symptom]
            else:
                hanbang = [1, 1, symptom]
            yangbang = [0, None, None]

        else:
            hanbang, yangbang = [None, None, None],[None, None, None]
    else:
        hanbang, yangbang = [None, None, None],[None, None, None]
            
    return hanbang, yangbang

        
```

```python
import re


def find_first_uncomfortable_symptom(text, word):
    # 선택적으로 괄호를 포함한 'word' 뒤에 나오는 첫 번째 '불편증상 여부' 패턴을 찾기 위한 정규 표현식
    pattern =  fr'\(?{word}\)?.*?불편증상 여부\s*:\s*(.*?)($|\n)'
    #pattern = fr'\(?{word}\)?.*?(?!{word2}).*?불편증상 여부\s*:\s*(.*?)($|\n)'
    # 텍스트에서 패턴을 검색
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        return match.group(1).strip()    
    return None
    
def find_first_uncomfortable_symptom2(text, word, word2):
    # 'word' 뒤에 나오는 첫 번째 '불편증상 여부' 패턴을 찾고, 그 안에 'word2'가 없는 경우를 고려한 정규 표현식
    pattern = fr'\(?{word}\)?.*?불편증상 여부\s*:\s*(.*?)($|\n)'
    
    # 텍스트에서 패턴을 검색
    matches = re.finditer(pattern, text, re.DOTALL)
    
    for match in matches:
        segment = match.group(0)
        print()
        print(match)
        print(segment)
        # 'word2'가 segment에 포함되지 않는 경우만 반환
        if word2 not in segment:
            print('성공')
            return match.group(1).strip()
    return None

# 예시 텍스트
text = '''
pattern1: ? - 한방 다이어트
? - 양방 식욕억제제
pattern2: 有
? - 한방 다이어트
 - 복용 기간(연/월) : 15년도 복용 (기간: 3개월~2개월) (제일 마지막) (감량: 거의 잘 안됨)
 - 불편증상 여부 : 두근거림 (심하진 않고 조금 어느정도 있었다.)

? - 양방 식욕억제제
 - 복용 기간(연/월) : 15년도 복용 (기간: 2~3개월) (감량: 운동도 병행해서 효과 양호 7-8kg)
 - 불편증상 여부 : 없음 / 불면
'''
result = find_first_uncomfortable_symptom2(text, '양방', '한방')
print(result) 

# 함수 실행 예시
result = find_first_uncomfortable_symptom2(text, '한방', '양방')
print(result) 
```

```python
import re

def find_info(df, i, n):
    global info_dict
    ProgressNote = f'ProgressNote_{n}'
    text = df.loc[i,ProgressNote]
    inbody = f'BMI_{n}'
    if n > 5 :
        print(i,n,"No match found")
        return
    if isinstance(text, float):
        print(i,n,"Float Error")
        find_info(df, i, n+1)
        return
    # 정규 표현식을 이용해 키, 체중, BMI 추출
    pattern = r'키\s*/\s*체중\s*/\s*BMI\s*:\s*(\d+(\.\d+)?)\s*/\s*(\d+(\.\d+)?)(?:\s*[>→\-~]\s*(\d+(\.\d+)?))?\s*/?\s*[^\d]*(\d+\.\d+)'
    pattern_coffee = r'커피\(증상\)\s*:\s*(.*?)(?:\n|$)'
    pattern_alcohol = r'술\(횟수\)\s*:\s*(.*?)(?:\n|$)'
    pattern_symptom = r'불편증상\s*:\s*(.*?)(?:\n|$)'
    
    matches = re.search(pattern, text)
    coffee_match = re.search(pattern_coffee, text)
    alcohol_match = re.search(pattern_alcohol, text)
    
    if matches:
        height = matches.group(1)
        weight = matches.group(5) if matches.group(5) else matches.group(3)
        bmi = matches.group(7)
        coffee = coffee_match.group(1).strip() if coffee_match else None
        alcohol = alcohol_match.group(1).strip() if alcohol_match else None
        diet_1, diet_2 = diet_patterns(text)
        info_dict['N']=n
        if np.isnan(df.loc[i,inbody]):
            info_dict['Visit'] = 2
        else:
            info_dict['Visit'] = 1
        info_dict['Height']=float(height)
        info_dict['Weight']=float(weight)
        info_dict['BMI']= float(bmi)
        info_dict['Coffee'] = coffee
        info_dict['Alcohol'] = alcohol
        info_dict['Diet1'] = diet_1
        info_dict['Diet2'] = diet_2
        
        one, two = diet_patterns(text)
        dtwo = diet_detail(two)
        done = diet_detail(one)
        if dtwo != ([None, None, None], [None, None, None]):
            detail = dtwo
        else:
            detail = done
        info_dict['한방 경험'] =detail[0][0]
        info_dict['한방 불편증상 여부'] = detail[0][1]
        info_dict['한방 불편증상 상세'] = detail[0][2]
        info_dict['양방 경험'] = detail[1][0]
        info_dict['양방 불편증상 여부'] = detail[1][1]
        info_dict['양방 불편증상 상세'] = detail[1][2]


        return
    else:
        find_info(df, i, n+1)
        print(i, n ,text)
        return
```
